### Soal 1 (15 poin)

#### Code

```python
# type your code here
df_india= pd.read_csv('./diabetes.csv')
dist_data = df_india[['Glucose','Insulin']]
print("Shape : ", dist_data.shape)
dist_data.head(10)
```

```python
dist_data.info()

fig = plt.figure(figsize=(20,6))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

dist_data.plot(kind='hist', alpha =0.35, ax=ax1)
sns.kdeplot(dist_data.Glucose, shade=True, ax=ax2)
sns.kdeplot(dist_data.Insulin, shade=True, ax=ax2)
ax1.set_xlabel("Value")
plt.xlabel("Value")
plt.suptitle("Distribusi variabel Glucose dan Insulin")
```

### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
# type your code here
#read data 
df_nasdaq= pd.read_csv('./Nasdaq.csv')
df_nikkei= pd.read_csv('./Nikkei.csv')
df_hangseng= pd.read_csv('./Hang_Seng.csv')


#pembulatan 2 angka dibelakang koma (,)
df_nasdaq = df_nasdaq.round(2)
df_nikkei = df_nikkei.round(2)
df_hangseng = df_hangseng.round(2)

#drop baris yang memiliki nilai null
df_nasdaq.dropna(inplace = True, axis = 0)
df_nikkei.dropna(inplace = True, axis = 0)
df_hangseng.dropna(inplace = True, axis = 0)

#ubah tipe data fitur date menjadi Date
df_nasdaq['Date'] = pd.to_datetime(df_nasdaq['Date'])
df_nikkei['Date'] = pd.to_datetime(df_nikkei['Date'])
df_hangseng['Date'] = pd.to_datetime(df_hangseng['Date'])

df_nasdaq.index = df_nasdaq["Date"]
df_nikkei.index = df_nikkei["Date"]
df_hangseng.index = df_hangseng["Date"]

df_nasdaq["Year"] = pd.DatetimeIndex(df_nasdaq['Date']).year
df_nasdaq = df_nasdaq.loc[df_nasdaq["Year"].isin(["2019"])]
df_nasdaq = df_nasdaq.drop(["Year"], axis=1)

df_nikkei["Year"] = pd.DatetimeIndex(df_nikkei['Date']).year
df_nikkei = df_nikkei.loc[df_nikkei["Year"].isin(["2019"])]
df_nikkei = df_nikkei.drop(["Year"], axis=1)

df_hangseng["Year"] = pd.DatetimeIndex(df_hangseng['Date']).year
df_hangseng = df_hangseng.loc[df_hangseng["Year"].isin(["2019"])]
df_hangseng = df_hangseng.drop(["Year"], axis=1)


#menambahkan kolom daily percententage change
df_nasdaq["Daily_Percentage_Change"] = df_nasdaq["Adj Close"].pct_change()*100
df_nikkei["Daily_Percentage_Change"] = df_nikkei["Adj Close"].pct_change()*100
df_hangseng["Daily_Percentage_Change"] = df_hangseng["Adj Close"].pct_change()*100

#menghapus baris data yang bernilai null
df_nasdaq.dropna(inplace = True, axis = 0)
df_nikkei.dropna(inplace = True, axis = 0)
df_hangseng.dropna(inplace = True, axis = 0)

#Plot daily percentage change
df_nasdaq["Daily_Percentage_Change"].plot(figsize = (17,10), fontsize = 12)
df_nikkei["Daily_Percentage_Change"].plot(figsize = (17, 10), fontsize = 12)
df_hangseng["Daily_Percentage_Change"].plot(figsize = (17, 10), fontsize = 12)
plt.legend(["Nasdaq","Nikkei","Hang Seng"], loc="upper right", fontsize="x-large")
plt.show()

def trend(x):
    if x> -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"

#menambahkan kolom trend
df_nasdaq["Trend"] = df_nasdaq["Daily_Percentage_Change"].apply(lambda x:trend(x))
df_nikkei["Trend"] = df_nikkei["Daily_Percentage_Change"].apply(lambda x:trend(x))
df_hangseng["Trend"] = df_hangseng["Daily_Percentage_Change"].apply(lambda x:trend(x))


#membuat pie chart frekuensi trend Nasdaq
df_nasdaq_pie = df_nasdaq.groupby("Trend")
df_nikkei_pie = df_nikkei.groupby("Trend")
pie_label = sorted([i for i in df_nasdaq.loc[:, "Trend"].unique()])
plt.pie(df_nasdaq_pie["Trend"].count(),
        labels = pie_label, 
        autopct = "%1.1f%%", 
        radius = 2,
        colors=["lightsalmon","lightcoral", "lightskyblue","aquamarine","khaki","mediumorchid", "blue", "green"]
        ) 
        
print('\n')
plt.title("Pie Chart Nasdaq")
plt.show()

#membuat pie chart frekuensi trend Nikkei
df_nikkei_pie = df_nikkei.groupby("Trend")
pie_label = sorted([i for i in df_nikkei.loc[:, "Trend"].unique()])
plt.pie(df_nikkei_pie["Trend"].count(),
        labels = pie_label, 
        autopct = "%1.1f%%", 
        radius = 2,
        colors=["lightsalmon","lightcoral", "lightskyblue","aquamarine","khaki","mediumorchid", "blue", "green"]
        ) 
        
print('\n')
plt.title("Pie Chart Nikkei")
plt.show()

#membuat pie chart frekuensi trend Hang Seng
df_hangseng_pie = df_hangseng.groupby("Trend")
pie_label = sorted([i for i in df_hangseng.loc[:, "Trend"].unique()])
plt.pie(df_hangseng_pie["Trend"].count(),
        labels = pie_label, 
        autopct = "%1.1f%%", 
        radius = 2,
        colors=["lightsalmon","lightcoral", "lightskyblue","aquamarine","khaki","mediumorchid", "blue", "green"]
        ) 
        
print('\n')
plt.title("Pie Chart Hangseng")
plt.show()


```

#### Penjelasan 
Berdasarkan plot dan chart di atas dapat dilihat bahwa:
1. Ketiga saham Nasdaq, Nikkei, dan Hangseng berdasarkan daily percentage change didominasi oleh hubungan Slight or No Change yaitu lebih dari 45% data time series. Sehingga pergerakkan ketiga saham tiap harinya banyak yang berubah sedikit saja (slight) atau bahkan diangap sebagai tidak ada perubahan.
2. Pada saham Nasdaq, Nikkei, Hangseng dengan frekuensi hubungan paling banyak kedua berdasarkan daily percentage change adalah Positive dengan 17,1%, Slight Negative dengan 14,6%, Slight Positive dengan 14,7% dari keseluruhan data time series. Saham Nikkei sangat berbeda dengan saham Nasdaq dan Hangseng yang cenderung Positif. 
3. Pada saham Nasdaq, Nikkei, Hangseng dengan frekuensi hubungan paling banyak ketiga berdasarkan daily percentage change adalah Slight Positive dengan 16,7%, Slight Positive dengan 13,3%, Positive dengan 13,1% dari keseluruhan data time series. 
4. Pada saham Nasdaq, Nikkei, Hangseng dengan frekuensi hubungan paling banyak keempat berdasarkan daily percentage change adalah Negative dan Slight Negative dengan 9,2%, Positive dengan 11,7%, Negative dengan 11,8% dari keseluruhan data time series.
5. Pada saham Nasdaq, Nikkei, Hangseng dengan frekuensi hubungan paling banyak kelima berdasarkan daily percentage change adalah Among top losers dengan 1,6%, Negative dengan 7,1%, Slight Negative dengan 11,4% dari keseluruhan data time series.
5. Pada saham Nasdaq, Nikkei, Hangseng dengan frekuensi hubungan paling banyak keenam berdasarkan daily percentage change adalah Among top gainers dengan 0,4%, Among top losers dengan 0,4%, Among top gainers dengan 0,4% dari keseluruhan data time series.

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# type your code here
df_gabung = pd.concat([df_nasdaq ["Adj Close"],df_nikkei["Adj Close"], df_hangseng["Adj Close"]], axis =1)
df_gabung.columns = ["Nasdaq","Nikkei","Hang_Seng"]
df_gabung.dropna(inplace=True, axis=0)

df_perc_chg = df_gabung.pct_change()*100
df_perc_chg.dropna(inplace = True, how="any", axis = 0)


df_gabung_vol = df_perc_chg[["Nasdaq","Nikkei","Hang_Seng"]].rolling(7).std()*np.sqrt(7)
df_gabung_vol.plot(figsize = (15,7))
plt.show()
```

#### Penjelasan 
Berdasarkan plot diatas dapat dilihat bahwa:
1. Fluktuasi untuk saham Nasdaq lebih banyak negative atau penurunan.
2. Fluktuasi untuk saham Nikkei yang paling banyak sering terjadi kenaikan dan penurunan sehingga cenderung tidak stabil.
3. Fluktuasi untuk saham Hangseng cenderung naik atau positif sehingga bisa dikatakan saham ini cukup stabil.
